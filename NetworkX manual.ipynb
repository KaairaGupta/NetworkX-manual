**To make a bipartite graph and add nodes in the two parts and join them with edges**

```python
Q=nx.Graph()
Q.add_nodes_from([1,2,3,4],bipartite=0)
Q.add_nodes_from(['a','b','c','d','e'],bipartite=1)
Q.add_edges_from([(1,'a'),(1,'b'),(2,'c'),(3,'c'),(4,'e'),(4,'d')])
```

```python
print(bipartite.is_bipartite(Q))
```

**To know if Q is bipartite or not**

```python
X=set([1,3])
bipartite.is_bipartite_node_set(Q,X)
```
